```python
st = pd.read_csv('./data/StudentsPerformance.csv')
```

## Scatter Plot 

```python
plt.figure(figsize=(8,8))
plt.xlabel('Reading Score')
plt.ylabel('Writing Score')
plt.title('Students Reading and Writing Scores')
N=1000
colors = np.random.rand(N)
plt.scatter(st['reading score'],st['writing score'],linewidths=2, c=colors)
```

```python
plt.figure(figsize=(8,8))
plt.xlabel('Reading Score')
plt.ylabel('Writing Score')
plt.title('Students Reading and Writing Scores')
plt.scatter(st['reading score'],st['writing score'],linewidths=2)
m, b = np.polyfit(st['reading score'], st['writing score'], 1)
plt.plot(st['reading score'], m*st['reading score'], + b)
```

```python
# you can see a positive relationship between reading and writing scores, obiously a student who is good in reading should be
# good in writing too
```
